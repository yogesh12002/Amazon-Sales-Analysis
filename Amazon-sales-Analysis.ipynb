# The highest purchase is for M-size

```python
plt.figure(figsize=(10,6))
sns.countplot(data=amazon_sales,x='Courier Status' , hue = 'Status')
plt.show()
```

# Courier is most common mode for shippment

```python
amazon_sales['Category']=amazon_sales['Category'].astype(str)
c_d=amazon_sales['Category']
plt.figure(figsize=(10,6))
plt.hist(c_d,bins=30,edgecolor="Red",color="green")
plt.xticks(rotation=90)
plt.show()
```

# T-shirts are highest selling product

```python
 check_B2B=amazon_sales['B2B'].value_counts()

plt.pie(check_B2B, labels=check_B2B.index, autopct='%1.1f%%')
plt.show()
```

# Maximum buyers are retailer 99.2% and  rest are wholesalers 0.8%

```python
x_data = amazon_sales['Category']
y_data = amazon_sales['Size']

plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Available size')
plt.show()
```

```python
top10_state=amazon_sales['ship-state'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.countplot(data=amazon_sales[amazon_sales['ship-state'].isin(top10_state.index)],x='ship-state')
plt.xlabel('State')
plt.ylabel('Order_count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

# most of the buyers are from Maharashtra
